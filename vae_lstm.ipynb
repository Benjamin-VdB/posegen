## VAE LSTM

```python
# defining the key parameters
batch_size = 256
input_dim = 26
latent_dim = 256
intermediate_dim = 128
epochs = 20
epsilon_std = 1.0
timesteps = 150
```

```python
samples.shape
```

```python
draw_sequence(samples[1500,:,:])
```

```python
def sampling(args):
    z_mean, z_log_sigma = args
    epsilon = K.random_normal(shape=(batch_size, latent_dim),
                              mean=0., stddev=epsilon_std)
    return z_mean + z_log_sigma * epsilon

def vae_loss(x, x_decoded_mean):
    xent_loss = objectives.mse(x, x_decoded_mean)
    kl_loss = - 0.5 * K.mean(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var))
    loss = xent_loss + kl_loss
    return loss
```

```python
# input to our encoder
x = Input(shape=(timesteps, input_dim))
# intermediate layer
e1 = CuDNNLSTM(intermediate_dim, return_sequences=True)(x)
e2 = CuDNNLSTM(latent_dim)(e1)
# defining the mean of the latent space
z_mean = Dense(latent_dim)(e2)
# defining the log variance of the latent space
z_log_var = Dense(latent_dim)(e2)
# note that "output_shape" isn't necessary with the TensorFlow backend
z = Lambda(sampling, output_shape=(latent_dim,))([z_mean, z_log_var])
# defining the encoder as a keras model
encoder = Model(x, [z_mean, z_log_var, z], name="encoder")
# print out summary of what we just did
encoder.summary()
```

```python
# Input to the decoder
input_decoder = Input(shape=(latent_dim,), name="decoder_input")
# taking the latent space to intermediate dimension
d1 = Dense(latent_dim)(input_decoder)
d2 = RepeatVector(timesteps)(d1)
d3 = CuDNNLSTM(intermediate_dim, return_sequences=True)(d2)
d4 = CuDNNLSTM(input_dim, return_sequences=True)(d3)

# getting the mean from the original dimension
#x_decoded = Dense(original_dim, activation='sigmoid', name="flat_decoded")(decoder_h)
# defining the decoder as a keras model
decoder = Model(input_decoder, d4, name="decoder")
decoder.summary()
```

```python
# grab the output. Recall, that we need to grab the 3rd element our sampling z
output_combined = decoder(encoder(x)[2])
# link the input and the overall output
vae = Model(x, output_combined)

vae.compile(optimizer='rmsprop', loss=vae_loss)
vae.summary()
```

```python
# training
vae.fit(samples[:-(len(samples)%batch_size)], samples[:-(len(samples)%batch_size)],
        shuffle=True,
        epochs=epochs,
        batch_size=batch_size)
```

```python
pred_samples = vae.predict(samples[:-(len(samples)%batch_size)], batch_size=batch_size)
```
